```python
# /content/drive/MyDrive/Colab Notebooks/20240204_Data/dm_export_IL_20000101_20231231.csv
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/20240204_Data/dm_export_Midwest-Cumulation-20000101_20231231.csv')
#df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/20240204_Data/dm_export_MO_20000101_20231231.csv')

df
```

```python
# Import necessary libraries
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
import time
import os
import csv
#import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, GRU, MaxPooling1D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from google.colab import files
import shutil #move saved files to Google Drive

# Redirect stdout to /dev/null
#os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # Suppress TensorFlow info and warning messages
#os.environ['PYTHONWARNINGS'] = 'ignore'   # Suppress Python warnings

SaveHistory = []
SavePrediction = []
SaveErr = []
target_prediction = 1 #set which one will be the prediction target (D0, D1, D3, etc. "1" refers to D1)
drive_folder = '/content/drive/MyDrive/Colab Notebooks/20240204_Data/Plot/20240730/Models'  # Specify the destination folder in your Google Drive

for i in [6]: #Missouri data: # 12 states
  DataSet1=df.iloc[:, i*6+2:i*6+8]
  print(DataSet1.head())

  X = DataSet1.iloc[:, :].values  # Include all columns
  #X = DataSet1.iloc[:, :-1].values  # Include all columns except the last one
  y = DataSet1.iloc[:, target_prediction].values #assign the column "D1" value to y
  DataTitle = DataSet1.iloc[:, target_prediction].name

  print(DataTitle, X.shape, y.shape)
  print(DataSet1.iloc[:, target_prediction].name)
  # Normalize the data using Min-Max scaling
  scaler = MinMaxScaler()
  X_scaled = scaler.fit_transform(X)

  # Define the sequence length (number of weeks to consider)
  sequence_length = 5

  # Prepare the data by creating sequences
  X_seq, y_seq = [], []
  for i in range(len(X_scaled) - sequence_length-1):
      X_seq.append(X_scaled[i:i+sequence_length])
      y_seq.append(y[i+sequence_length+1])

  X_seq, y_seq = np.array(X_seq), np.array(y_seq)

  # Split the data into training and testing sets
  # X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, train_size=0.75, test_size=0.1, random_state=42)
  X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.15, shuffle = False)

  #X_train, y_train = X_seq[:1000], y_seq[:1000]
  #X_val, y_val = X_seq[1000:1140], y_seq[1000:1140]
  #X_test, y_test = X_seq[1140:], y_seq[1140:]

  print('X_train.shape, y_train.shape, X_test.shape, y_test.shape ')
  print(X_train.shape, y_train.shape, X_test.shape, y_test.shape) #X_val.shape, y_val.shape,

  set_epochs =10

  # Build the Conv1D model ============================================================
  model = Sequential()
  model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(sequence_length, X_train.shape[2])))
  model.add(MaxPooling1D(pool_size=2))
  model.add(Flatten())
  model.add(Dense(50, activation='relu'))
  model.add(Dense(1))  # No activation function for regression task

  # Compile the model
  optimizer = Adam(learning_rate=0.001)  # Adjust learning rate as needed
  model.compile(optimizer='adam', loss='mean_squared_error')  # Change loss based on your task

  # Implement learning rate reduction callback
  reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)

  # Train the model
  #history = model.fit(X_train, y_train, epochs=40, batch_size=32, validation_data=(X_test, y_test))
  history = model.fit(X_train, y_train, epochs=set_epochs, batch_size=32, validation_data=(X_test, y_test), callbacks=[reduce_lr])

  # Save history data into list
  SaveHistory.append([f"{DataTitle}_Conv1D_loss"] + history.history['loss'])
  SaveHistory.append([f"{DataTitle}_Conv1D_Val_loss"] + history.history['val_loss'])

  # Plot training history
  plt.plot(history.history['loss'], label='Training Loss')
  plt.plot(history.history['val_loss'], label='Validation Loss')
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  plt.title(f'{DataSet1.iloc[:, target_prediction].name} {sequence_length} weeks Conv1D model Training and Validation Loss')
  plt.legend()

  saveFileName = f'{DataSet1.iloc[:, target_prediction].name}_{sequence_length}_weeks_Conv1D_{set_epochs}_epochs_Training_and_Validation_Loss.png'

  plt.savefig(saveFileName)
  shutil.move(f'/content/{saveFileName}', drive_folder)  # Move the file to your Google Drive folder
  time.sleep(0.3)

  plt.show()
  time.sleep(0.3)

  # Make predictions
  predictions = model.predict(X_test)

  # Save prediction data into list
  SavePrediction.append([f'{DataTitle}_Conv1D_Actual'] + list(y_test))
  SavePrediction.append([f'{DataTitle}_Conv1D_Prediction'] + list(predictions))

  # Accuracy of predictions
  prediction1 = predictions.reshape(-1)  # Selecting values and reshaping
  # Calculate Root Mean Squared Error (RMSE)
  rmse = np.sqrt(mean_squared_error(y_test, prediction1))

  # Calculate Mean Absolute Error (MAE)
  mae = mean_absolute_error(y_test, prediction1)

  # Save error data into list
  SaveErr.append([f'{DataTitle}_Conv1D_RMSE'] + [rmse])
  SaveErr.append([f'{DataTitle}_Conv1D_MAE'] + [mae])

  print("Root Mean Squared Error (RMSE):", rmse)
  print("Mean Absolute Error (MAE):", mae)
  # accuracy = accuracy_score(y_test, prediction1)
  #print("Accuracy: ", accuracy)

  #predictions, y_test
  plt.plot(y_test, label='Actual Data')
  plt.plot(predictions, label='Prediction')
  plt.xlabel('Weeks')
  plt.ylabel(DataSet1.iloc[:, target_prediction].name)
  plt.title(f"{sequence_length} weeks data predict {DataSet1.iloc[:, target_prediction].name} using Conv1D model with {set_epochs} epochs")
  plt.legend()

  saveFileName = f'{DataSet1.iloc[:, target_prediction].name}_{sequence_length}_weeks_Conv1D_{set_epochs}_epochs.png'
  plt.savefig(saveFileName)
  shutil.move(f'/content/{saveFileName}', drive_folder)  # Move the file to your Google Drive folder
  time.sleep(0.3)

  plt.show()
  time.sleep(0.3)

  # Build the LSTM model=======================================================
  model1 = Sequential()
  model1.add(LSTM(units=64, activation='relu', input_shape=(sequence_length, X_train.shape[2])))
  model1.add(Dense(50, activation='relu'))
  model1.add(Dense(1))  # No activation function for regression task

  # Compile the model
  model1.compile(optimizer='adam', loss='mean_squared_error')
  # Train the model
  history = model1.fit(X_train, y_train, epochs=set_epochs, batch_size=32, validation_data=(X_test, y_test))

  # Save history data into list
  SaveHistory.append([f"{DataTitle}_LSTM_loss"] + history.history['loss'])
  SaveHistory.append([f"{DataTitle}_LSTM_Val_loss"] + history.history['val_loss'])

  # Plot training history
  plt.plot(history.history['loss'], label='Training Loss')
  plt.plot(history.history['val_loss'], label='Validation Loss')
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  plt.title(f'{DataSet1.iloc[:, target_prediction].name} {sequence_length} weeks LSTM model Training and Validation Loss')
  plt.legend()

  saveFileName = f'{DataSet1.iloc[:, target_prediction].name}_{sequence_length}_weeks_LSTM_{set_epochs}_epochs_Training_and_Validation_Loss.png'

  plt.savefig(saveFileName)
  shutil.move(f'/content/{saveFileName}', drive_folder)  # Move the file to your Google Drive folder
  time.sleep(0.3)
  plt.show()
  time.sleep(0.3)

  # Make predictions
  predictions = model1.predict(X_test)

  # Save prediction data into list
  SavePrediction.append([f'{DataTitle}_LSTM_Actual'] + list(y_test))
  SavePrediction.append([f'{DataTitle}_LSTM_Prediction'] + list(predictions))

  # Accuracy of predictions
  prediction1 = predictions.reshape(-1)  # Selecting values and reshaping
  # Calculate Root Mean Squared Error (RMSE)
  rmse = np.sqrt(mean_squared_error(y_test, prediction1))

  # Calculate Mean Absolute Error (MAE)
  mae = mean_absolute_error(y_test, prediction1)

  # Save error data into list
  SaveErr.append([f'{DataTitle}_LSTM_RMSE'] + [rmse])
  SaveErr.append([f'{DataTitle}_LSTM_MAE'] + [mae])

  print("Root Mean Squared Error (RMSE):", rmse)
  print("Mean Absolute Error (MAE):", mae)
  # accuracy = accuracy_score(y_test, prediction1)
  #print("Accuracy: ", accuracy)

  #Plot predictions, y_test
  plt.plot(y_test, label='Actual Data')
  plt.plot(predictions, label='Prediction')
  plt.xlabel('Weeks')
  plt.ylabel(DataSet1.iloc[:, target_prediction].name)
  plt.title(f"{sequence_length} weeks data predict {DataSet1.iloc[:, target_prediction].name} using LSTM model with {set_epochs} epochs")
  plt.legend()

  saveFileName = f'{DataSet1.iloc[:, target_prediction].name}_{sequence_length}_weeks_LSTM_{set_epochs}_epochs.png'
  plt.savefig(saveFileName)
  shutil.move(f'/content/{saveFileName}', drive_folder)  # Move the file to your Google Drive folder
  time.sleep(0.3)
  plt.show()
  time.sleep(0.3)

  # Build the GRU model =========================================================================
  model2 = Sequential()
  model2.add(GRU(units=64, activation='relu', input_shape=(sequence_length, X_train.shape[2])))
  model2.add(Dense(50, activation='relu'))
  model2.add(Dense(1))  # No activation function for regression task

  # Compile the model
  model2.compile(optimizer='adam', loss='mean_squared_error')

  # Train the model
  history = model2.fit(X_train, y_train, epochs=set_epochs, batch_size=32, validation_data=(X_test, y_test))

  # Save history data into list
  SaveHistory.append([f"{DataTitle}_GRU_loss"] + history.history['loss'])
  SaveHistory.append([f"{DataTitle}_GRU_Val_loss"] + history.history['val_loss'])

  # Plot training history
  plt.plot(history.history['loss'], label='Training Loss')
  plt.plot(history.history['val_loss'], label='Validation Loss')
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  plt.title(f'{DataSet1.iloc[:, target_prediction].name} {sequence_length} weeks GRU model Training and Validation Loss')
  plt.legend()

  saveFileName = f'{DataSet1.iloc[:, target_prediction].name}_{sequence_length}_weeks_GRU_{set_epochs}_epochs_Training_and_Validation_Loss.png'

  plt.savefig(saveFileName)
  shutil.move(f'/content/{saveFileName}', drive_folder)  # Move the file to your Google Drive folder
  time.sleep(0.3)
  plt.show()
  time.sleep(0.3)

  # Make predictions
  predictions = model2.predict(X_test)

  # Save prediction data into list
  SavePrediction.append([f'{DataTitle}_GRU_Actual'] + list(y_test))
  SavePrediction.append([f'{DataTitle}_GRU_Prediction'] + list(predictions))

  # Accuracy of predictions
  prediction1 = predictions.reshape(-1)  # Selecting values and reshaping
  # Calculate Root Mean Squared Error (RMSE)
  rmse = np.sqrt(mean_squared_error(y_test, prediction1))

  # Calculate Mean Absolute Error (MAE)
  mae = mean_absolute_error(y_test, prediction1)

  # Save error data into list
  SaveErr.append([f'{DataTitle}_GRU_RMSE'] + [rmse])
  SaveErr.append([f'{DataTitle}_GRU_MAE'] + [mae])

  print("Root Mean Squared Error (RMSE):", rmse)
  print("Mean Absolute Error (MAE):", mae)
  # accuracy = accuracy_score(y_test, prediction1)
  #print("Accuracy: ", accuracy)

  #Plot predictions, y_test
  plt.plot(y_test, label='Actual Data')
  plt.plot(predictions, label='Prediction')
  plt.xlabel('Weeks')
  plt.ylabel(DataSet1.iloc[:, target_prediction].name)
  plt.title(f"{sequence_length} weeks data predict {DataSet1.iloc[:, target_prediction].name} using GRU model with {set_epochs} epochs")
  plt.legend()

  saveFileName = f'{DataSet1.iloc[:, target_prediction].name}_{sequence_length}_weeks_GRU_{set_epochs}_epochs.png'
  plt.savefig(saveFileName)
  shutil.move(f'/content/{saveFileName}', drive_folder)  # Move the file to your Google Drive folder
  time.sleep(0.3)
  plt.show()
  time.sleep(0.3)


  # =======Save training and prediction data=============================================
  drive_folder2 = drive_folder + '/DATA'  # Specify the destination folder in your Google Drive

  csv_file = 'MO_5_weeks_10_epochs_Models_Training_Loss.csv'
  try:
      with open(csv_file, mode='w', newline='') as file:
          writer = csv.writer(file)
          writer.writerows(SaveHistory)
  except IOError:
      print("An error occurred while writing the CSV file.")

  shutil.move(f'/content/{csv_file}', drive_folder2)  # Move the file to your Google Drive folder
  # Download the CSV file to the local drive
  #from google.colab import files
  #files.download(csv_file)

  csv_file = 'MO_5_weeks_10_epochs_Models_Prediction.csv'
  try:
      with open(csv_file, mode='w', newline='') as file:
          writer = csv.writer(file)
          writer.writerows(SavePrediction)
  except IOError:
      print("An error occurred while writing the CSV file.")

  shutil.move(f'/content/{csv_file}', drive_folder2)  # Move the file to your Google Drive folder
  # Download the CSV file to the local drive
  #from google.colab import files
  #files.download(csv_file)

  csv_file = 'MO_5_weeks_10_epochs_Models_Err.csv'
  try:
      with open(csv_file, mode='w', newline='') as file:
          writer = csv.writer(file)
          writer.writerows(SaveErr)
  except IOError:
      print("An error occurred while writing the CSV file.")

  shutil.move(f'/content/{csv_file}', drive_folder2)  # Move the file to your Google Drive folder
  # Download the CSV file to the local drive
  #from google.colab import files
  #files.download(csv_file)
```
